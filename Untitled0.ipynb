```
def plot_charts(dataset, title):
  fig, axs = plt.subplots(2, 1, figsize=(25, 10))

  # A/B rates
  axs[0].plot(dataset.index, dataset['A_rate'], label='A_rate')
  axs[0].plot(dataset.index, dataset['B_rate'], label='B_rate')
  axs[0].set_ylabel('Rate')
  axs[0].legend()
  axs[0].set_title('Rate ' + title)
  axs[0].grid(True)

  # light parts
  axs[1].plot(dataset.index, dataset['A_CH4'], label='A_CH4')
  axs[1].plot(dataset.index, dataset['A_C2H6'], label='A_C2H6')
  axs[1].plot(dataset.index, dataset['A_C3H8'], label='A_C3H8')
  axs[1].plot(dataset.index, dataset['A_iC4H10'], label='A_iC4H10')
  axs[1].plot(dataset.index, dataset['A_nC4H10'], label='A_nC4H10')
  axs[1].plot(dataset.index, dataset['A_iC5H12'], label='A_iC5H12')
  axs[1].plot(dataset.index, dataset['A_nC5H12'], label='A_nC5H12')
  axs[1].plot(dataset.index, dataset['A_C6H14'], label='A_C6H14')
  axs[1].set_ylabel('Share')
  axs[1].legend()
  axs[1].set_title('Parts ' + title)
  axs[1].grid(True)
```

We need a filtering and filling function

```
# filling nans with SARIMAX
#from statsmodels.tsa.statespace.sarimax import SARIMAX
#from sklearn.metrics import mean_squared_error

def filling_nans(in_timeserie, window=96, variance=2):
  timeserie = in_timeserie.copy()

  for i in range(window, len(timeserie)):
    current_index = timeserie.index[i]

    train = timeserie[timeserie.index[i - window] : timeserie.index[i - 1]]
    train = train.fillna(train.mean())

    #s_model = SARIMAX(train, validate_specification=False)
    #results = s_model.fit()
    #predict = results.predict()
    #forecast = results.forecast(1)
    #rmse = mean_squared_error(train.fillna(train.mean()), predict)

    if timeserie[current_index] != timeserie[current_index]:
      timeserie[current_index] = train[train.last_valid_index()]
    elif timeserie[current_index] > timeserie[train.last_valid_index()] + variance: #3 * rmse:
      timeserie[current_index] = train[train.last_valid_index()]
    elif timeserie[current_index] < timeserie[train.last_valid_index()] - variance: #3 * rmse:
      timeserie[current_index] = train[train.last_valid_index()]

  return timeserie

print('Compiling filling_nans - OK')
```

```
# Fixing X_train series
X_train_d = X_train.copy()
X_train_d['A_CH4'] = filling_nans(X_train_d['A_CH4'], window=48, variance=2)
X_train_d['A_C2H6'] = filling_nans(X_train_d['A_C2H6'], window=48, variance=2)
X_train_d['A_C3H8'] = filling_nans(X_train_d['A_C3H8'], window=48, variance=5)
X_train_d['A_iC4H10'] = filling_nans(X_train_d['A_iC4H10'], window=48, variance=2)
X_train_d['A_nC4H10'] = filling_nans(X_train_d['A_nC4H10'], window=48, variance=2)
X_train_d['A_iC5H12'] = filling_nans(X_train_d['A_iC5H12'], window=48, variance=2)
X_train_d['A_nC5H12'] = filling_nans(X_train_d['A_nC5H12'], window=48, variance=2)
X_train_d['A_C6H14'] = filling_nans(X_train_d['A_C6H14'], window=48, variance=2)
X_train_d['A_rate'] = filling_nans(X_train_d['A_rate'], window=48, variance=20)
X_train_d['B_rate'] = filling_nans(X_train_d['B_rate'], window=48, variance=20)
print('Fixing X_train - OK')
```

```
# adding lags into features
def adding_lags(dataset):
  for col_name in dataset.columns:
    for i in range (1, 12):
      dataset[col_name + '_{:d}'.format(i)] = dataset[col_name].shift(i)
  return dataset
```

MAPE scorer

```
#from sklearn.utils import check_arrays
def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```
# B_C2H6
print('starting B_C2H6...')
model_B_C2H6 = CatBoostRegressor(loss_function='MAE', iterations=100)
model_B_C2H6.fit(X_train_d, Y_train_d['B_C2H6'], verbose=1000)
print('finished B_C2H6')

# B_C3H8
print('starting B_C3H8...')
model_B_C3H8 = CatBoostRegressor(loss_function='MAE', iterations=100)
model_B_C3H8.fit(X_train_d, Y_train_d['B_C3H8'], verbose=1000)
print('finished B_C3H8')

# B_iC4H10
print('starting B_iC4H10...')
model_B_iC4H10 = CatBoostRegressor(loss_function='MAE', iterations=100)
model_B_iC4H10.fit(X_train_d, Y_train_d['B_iC4H10'], verbose=1000)
print('finished B_iC4H10')

# B_nC4H10
print('starting B_nC4H10...')
model_B_nC4H10 = CatBoostRegressor(loss_function='MAE', iterations=100)
model_B_nC4H10.fit(X_train_d, Y_train_d['B_nC4H10'], verbose=1000)
print('finished B_nC4H10')
```

Evaluating train set

```
# train MAPE
B_C2H6_train_predict = model_B_C2H6.predict(X_train_d)
print('Train B_C2H6 MAPE:', mean_absolute_percentage_error(Y_train_d['B_C2H6'], B_C2H6_train_predict))

B_C3H8_train_predict = model_B_C3H8.predict(X_train_d)
print('Train B_C3H8 MAPE:', mean_absolute_percentage_error(Y_train_d['B_C3H8'], B_C3H8_train_predict))

B_iC4H10_train_predict = model_B_iC4H10.predict(X_train_d)
print('Train B_iC4H10 MAPE:', mean_absolute_percentage_error(Y_train_d['B_iC4H10'], B_iC4H10_train_predict))

B_nC4H10_train_predict = model_B_nC4H10.predict(X_train_d)
print('Train B_nC4H10 MAPE:', mean_absolute_percentage_error(Y_train_d['B_nC4H10'], B_nC4H10_train_predict))
```
